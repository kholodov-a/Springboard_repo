```python
from ultralytics import YOLO
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the YOLOv8 model (classification mode)
model = YOLO('../Models_copy/yolo_cls/yolov8s-cls.pt')

# Train the model
model.train(
    data = "../Data_copy/processed/YOLO",    # Path to data folders
    epochs = 30,                        # Number of training epochs
    imgsz = 224,                        # Input image size (resize images to 224x224)
    batch = 32,                         # Batch size
    project = "../Models_copy/yolo_cls",     # Project folder for saving results
    name = "final_model",               # Name of the training run
    device="mps" )                      # Use the MPS backend for GPU
```

```python
# Evaluate the model
metrics = model.val()
print("Validation Results:", metrics)
```

```python
# Predict on the cardboard images for demostration
path = '/Users/a.kholodov/Documents/02. Personal/20. Education/50. Universities/Springboard/23. Deep Learning/23.4 Capstone Three - Project Ideas and Proposal/Data_copy/processed/YOLO_test'
model = YOLO('../Models_copy/yolo_cls/final_model/weights/best.pt')
results = model.predict(source = path, save = True, verbose = False)
```

```python
%matplotlib inline
images = np.random.choice(range(len(results)), 16, replace = False) 

fig, ax = plt.subplots(figsize = (12, 12), nrows = 4, ncols = 4)
for i in range(16):
    ax[i // 4, i % 4].imshow(results[images[i]].orig_img.astype('uint8'))
    ax[i // 4, i % 4].set_title(os.path.basename(results[images[i]].path) + '\n' + model.names[results[images[i]].probs.top1])
    ax[i // 4, i % 4].axis("off")

plt.show()
```
